### Pertanyaan 5:

```python
day = pd.read_csv('day.csv')

# MENGAMBIL DATA SUHU & CNT
x = day['temp']     # suhu (skala 0–1)
y = day['cnt']      # jumlah penyewa

# 3. MEMBUAT TREND LINE
m, b = np.polyfit(x, y, 1)
trend_y = m*x + b

# MEMBUAT PLOT
plt.figure(figsize=(10,6))

plt.scatter(x, y)                            # titik data
plt.plot(x, trend_y, color='orange', 
         linewidth=2)                        # garis tren oranye


plt.title("HUBUNGAN SUHU & PENYEWA")
plt.xlabel("Suhu (temp, normalisasi 0-1)")
plt.ylabel("Penyewa (cnt)")

plt.grid(True)
plt.show()
```


## Analisis Korelasi Suhu-Permintaan
Scatter plot di atas dengan garis tren linear mendemonstrasikan hubungan positif yang jelas antara suhu ternormalisasi (temp) dan jumlah penyewa (cnt), meskipun dengan tingkat penyebaran yang cukup signifikan.

- Karakteristik Hubungan:
Korelasi positif moderat (sekitar 0,4-0,6) terlihat dari pola titik yang secara umum bergerak dari kiri bawah ke kanan atas
Garis tren menunjukkan bahwa setiap kenaikan suhu ternormalisasi dikaitkan dengan peningkatan jumlah penyewa
Pada rentang suhu rendah (0,05-0,30), sebagian besar titik terkonsentrasi di bagian bawah grafik dengan jumlah penyewa umumnya di bawah 3.000
Pada rentang suhu menengah hingga tinggi (0,40-0,70), titik-titik menyebar lebih luas dengan banyak observasi mencapai 5.000-8.000 penyewa
Terdapat variabilitas substansial pada setiap level suhu, menunjukkan bahwa suhu bukan satu-satunya faktor penentu

- Penjelasan Pola Non-Linear:
Meskipun tren umum positif, hubungan sebenarnya cenderung non-linear:
Pada suhu sangat rendah (mendekati 0), hampir tidak ada yang bersepeda terlepas dari faktor lain
Pada rentang suhu nyaman (sekitar 0,40-0,70 atau 15-27°C dalam skala asli), terjadi lonjakan permintaan yang substansial
Pada suhu sangat tinggi (mendekati 1,0 atau di atas 35°C), penelitian menunjukkan bahwa permintaan cenderung menurun kembali karena ketidaknyamanan, meskipun jumlah observasi di rentang ini terbatas dalam dataset


- Faktor Moderasi:
Variabilitas besar pada scatter plot menunjukkan bahwa pengaruh suhu dimoderasi oleh faktor lain:
Kelembapan dan cuaca: hari dengan suhu tinggi namun hujan atau kelembapan tinggi akan memiliki penyewa lebih rendah dari yang diprediksi suhu saja
Musim: suhu 0,5 pada musim semi berbeda dampaknya dengan suhu 0,5 pada musim gugur karena ekspektasi cuaca yang berbeda
Hari kerja vs libur: pada suhu yang sama, hari kerja cenderung memiliki pola berbeda (dua puncak) dibanding akhir pekan
Kecepatan angin: angin kencang dapat mengurangi kenyamanan bersepeda meskipun suhu ideal

- Implikasi Prediktif:
Untuk model prediksi yang akurat, suhu harus dikombinasikan dengan fitur lain dalam model multivariat. Model yang hanya menggunakan suhu akan memiliki error prediksi tinggi (ditunjukkan oleh R² yang tidak terlalu tinggi pada regresi linear sederhana). Namun, suhu tetap merupakan salah satu prediktor paling penting dan harus selalu disertakan dalam model forecasting permintaan bike sharing.

```python
day = pd.read_csv('day.csv')

# 4. Membuat histogram distribusi suhu
plt.figure(figsize=(8, 5))

plt.hist(day['temp'],
         bins=30,# jumlah kotak histogram
         color='blue',
         edgecolor='blue',
         alpha=0.7)

# 5. Menambahkan garis mean dan median
mean_temp = day['temp'].mean()
median_temp = day['temp'].median()

plt.axvline(mean_temp,
            color='red',
            linestyle='--',
            linewidth=2,
            label=f'Mean = {mean_temp:.3f}')

plt.axvline(median_temp,
            color='green',
            linestyle='--',
            linewidth=2,
            label=f'Median = {median_temp:.3f}')

# 6. Memberi judul dan label sumbu
plt.title('Distribusi Suhu (temp) Sepanjang Pengamatan',
          fontsize=14, fontweight='bold')
plt.xlabel('Suhu ternormalisasi (temp)', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)

plt.grid(axis='y', alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```

Musim panas dan gugur memiliki penyewaan tertinggi: kedua musim ini bertepatan dengan konsentrasi suhu di rentang optimal (0,4-0,7)
Musim dingin memiliki penyewaan terendah: sebagian besar hari musim dingin berada di ekor kiri distribusi (suhu di bawah 0,35) yang kurang mendukung aktivitas bersepeda


- Implikasi untuk Manajemen Sistem:
Dengan mengetahui bahwa sebagian besar waktu operasional (sekitar 70-80% hari) berada pada rentang suhu 0,3-0,7, sistem dapat:
Mengoptimalkan kapasitas untuk kondisi "normal" (suhu menengah) sebagai baseline operasional
Mempersiapkan protokol khusus untuk hari dengan suhu ekstrem rendah (di bawah 0,2) dimana permintaan akan sangat rendah dan pemeliharaan bisa diprioritaskan
Merencanakan kampanye promosi berdasarkan prediksi cuaca, memanfaatkan hari-hari dengan suhu ideal (0,45-0,65) untuk menarik pengguna baru

- Kesimpulan Komprehensif:
Keenam analisis grafik di atas secara kolektif mengungkapkan bahwa sistem bike sharing dipengaruhi oleh interaksi kompleks antara faktor temporal (jam, hari, musim), faktor meteorologi (suhu, cuaca), dan faktor perilaku pengguna (komuter vs rekreasional). Distribusi yang condong kanan pada level per jam, pola musiman yang kuat pada level harian, perbedaan hari kerja-akhir pekan, variasi drastis antar musim, korelasi positif dengan suhu, dan konsentrasi suhu di rentang optimal semuanya menunjukkan bahwa perencanaan operasional yang efektif memerlukan pendekatan multidimensional yang mempertimbangkan semua faktor ini secara simultan. Model prediktif yang baik harus mengintegrasikan semua variabel ini untuk memberikan forecast yang akurat dan mendukung keputusan alokasi sumber daya yang optimal.

Distribusi jumlah penyewa per jam (cnt) sangat miring ke kanan (right-skewed). Ini berarti mayoritas jam pengamatan memiliki jumlah penyewa yang sangat sedikit (puncak frekuensi berada mendekati 0). Meskipun demikian, ada sejumlah kecil jam yang mencatat jumlah penyewa yang sangat tinggi.
